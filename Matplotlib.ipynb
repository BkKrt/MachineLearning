```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
```

```python
x = [1,2,3,4]
y = [1,4,9,16]

plt.plot(x,y)
plt.show()
```

```python
plt.title("ilk grafiğimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show()
```

```python
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16])
plt.title("ilk grafiğimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, label="x^2", color ="green")
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16,25])
plt.title("ilk grafiğimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
plt.plot(x,y, label="x^2", color ="blue", linewidth = 2, linestyle = "--",marker="*")
plt.xticks([1,2,3,4])
plt.yticks([1,4,9,16,25])
plt.title("ilk grafiğimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

```python
plt.plot(x,y, label="x^2", color ="blue", linewidth = 2, linestyle = "--",marker="*")
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,1,4,9,16,25])
plt.title("ilk grafiğimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")


x2 = np.arange(0,5,0.5)
plt.plot(x2,x2*2,color ="red", linewidth = 2, marker = "*",label="2*x")

plt.legend()
plt.show()
```

```python
plt.plot(x,y, label="x^2", color ="blue", linewidth = 2, linestyle = "--",marker="*")
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,1,4,9,16,25])
plt.title("ilk grafiğimiz")
plt.xlabel("x axis")
plt.ylabel("y axis")


x2 = np.arange(0,5,0.5)
plt.plot(x2,x2*2,color ="red", linewidth = 2, marker = "*",label="2*x")

plt.legend()
plt.savefig("ilkgrafiğimiz.png" , dpi =300)
plt.show()
```

```python
x = ["Ankara" , "İstanbul", "İzmir"]
y = [120, 178, 87]
plt.bar(x,y)
plt.show()
```

```python
x = ["Ankara" , "İstanbul", "İzmir"]
y = [120, 178, 87]

cubuklar = plt.bar(x,y)
cubuklar[2].set_hatch(".")
cubuklar[1].set_hatch("/")
cubuklar[0].set_hatch("*")
plt.show()
```

```python
gas = pd.read_csv("petrolfiyatlari.csv")

gas
```

```python
plt.title("Petrol Fiyatları")
plt.plot(gas["Year"], gas["USA"], "r-", label = "USA")

plt.xlabel("Yıl")
plt.ylabel("Dolar")

plt.legend()
plt.show()
```

```python
plt.title("Petrol Fiyatları")
plt.plot(gas["Year"], gas["UK"], "b-", label = "UK")

plt.xlabel("Yıl")
plt.ylabel("Dolar")

plt.legend()
plt.show()
```

```python
plt.title("Petrol Fiyatları")
plt.plot(gas.Year, gas.USA, "b.-", label = "USA")
plt.plot(gas.Year, gas.Canada,"r.-", label = "Kanada")
plt.plot(gas.Year, gas["South Korea"], "g.-", label ="Güney Kore")
plt.plot(gas.Year, gas.France,"y.-", label = "Fransa")
plt.plot(gas.Year, gas.Germany,color ="black",marker = ".",label = "Almanya")
plt.xlabel("Yıl")
plt.ylabel("Dolar")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.title("Petrol Fiyatları")
plt.plot(gas.Year, gas.USA, "b.-", label = "USA")
plt.plot(gas.Year, gas.Canada,"r.-", label = "Kanada")
plt.plot(gas.Year, gas["South Korea"], "g.-", label ="Güney Kore")
plt.plot(gas.Year, gas.France,"y.-", label = "Fransa")
plt.plot(gas.Year, gas.Germany,color ="black",marker = ".",label = "Almanya")
plt.xlabel("Yıl")
plt.ylabel("Dolar")

plt.legend()
plt.savefig("PetrolFiyatları.png", dpi = 300)
plt.show()
```
